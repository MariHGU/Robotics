```python
# 1st option: Numpy
import numpy as np
# np.arccos, np.trace, np.clip to avoid NaN, np.sin

R = np.array([[0, 0, -1],
     [-1, 0, 0],
     [0, 1, 0]])

theta = np.arccos(np.clip((np.trace(R)-1)/2, -1.0, 1.0))
omega = np.array([R[2, 1] - R[1, 2], R[0, 2] - R[2, 0], R[1, 0] - R[0, 1]]) / (2 * np.sin(theta)) if np.sin(theta) != 0 else np.array([0, 0, 0])
omega_theta = omega*theta

print(f'Omega and theta: {omega_theta}')
```

```python
# 2nd option: Pinocchio
import pinocchio as pin
import numpy as np

R = np.array([[0, 0, -1],
     [-1, 0, 0],
     [0, 1, 0]])

# finding omega and theta using pinocchios built in log function
omega_theta = pin.log3(R)

print("Omega and Theta: ", omega_theta)
```
